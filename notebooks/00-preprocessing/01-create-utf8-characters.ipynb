```python
data = pd.read_csv('dataset-string-similarity.txt', sep='\t', header=None)
```

```python
data.head()
```

```python
data.columns = ['name_1', 'name_2', 'match', 'id_1', 'id_2', 'alphabet_1', 'alphabet_2', 'language_1', 'language_2']
```

```python
df_latin_pairs = data[(data.alphabet_1 == 'LATIN') & (data.alphabet_2 == 'LATIN')].dropna()
```

```python
df_latin_pairs.groupby('match').count()
```

```python
df_latin_pairs
```

```python
from Levenshtein import jaro_winkler
```

```python
jaro_winkler('home', 'home 2')
```

```python
df_latin_pairs['jaro_winkler'] = df_latin_pairs.apply(lambda row: jaro_winkler(row['name_1'], row['name_2']), axis=1)
```

```python
df_latin_pairs_filtered = df_latin_pairs[df_latin_pairs['jaro_winkler']>.75]
```

```python
df_latin_pairs_filtered.groupby('match').count()
```
